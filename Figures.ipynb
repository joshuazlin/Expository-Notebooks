```python
import matplotlib.pyplot as plt
import numpy as np
import itertools
```

# Cluster Algorithm

```python
#Figure 1
plt.figure(figsize=(18,6))

np.random.seed(45)
plt.subplot(1, 3, 1)
c1 = plt.Circle((0, 0), 0.1, facecolor='w',edgecolor='k')
plt.gca().add_patch(c1)
plt.grid(color='gray')
for axi in (plt.gca().xaxis, plt.gca().yaxis):
    for tic in axi.get_major_ticks():
        tic.tick1On = tic.tick2On = False
        tic.label1On = tic.label2On = False

for (x,y) in itertools.product(np.linspace(-1,1,9),repeat=2):
    if np.random.randint(2):
        plt.arrow(x,y-0.05,0,0.05,width=0.015,color='blue')
    else:
        plt.arrow(x,y+0.05,0,-0.05,width=0.015,color='red')
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.title("Pick a seed site randomly")

np.random.seed(45)
plt.subplot(1, 3, 2)
c1 = plt.Circle((0, 0), 0.1, facecolor='w',edgecolor='k')
c2 = plt.Circle((0.25, 0), 0.1, facecolor='w',edgecolor='gray')
c3 = plt.Circle((-0.25, 0), 0.1, facecolor='w',edgecolor='gray')
c4 = plt.Circle((-0.25, 0.25), 0.1, facecolor='w',edgecolor='gray')
c5 = plt.Circle((0, -0.25), 0.1, facecolor='w',edgecolor='gray')
plt.gca().add_patch(c1);plt.gca().add_patch(c2);plt.gca().add_patch(c3);
plt.gca().add_patch(c4);plt.gca().add_patch(c5)
plt.grid(color='gray')
for axi in (plt.gca().xaxis, plt.gca().yaxis):
    for tic in axi.get_major_ticks():
        tic.tick1On = tic.tick2On = False
        tic.label1On = tic.label2On = False

for (x,y) in itertools.product(np.linspace(-1,1,9),repeat=2):
    if np.random.randint(2):
        plt.arrow(x,y-0.05,0,0.05,width=0.015,color='blue')
    else:
        plt.arrow(x,y+0.05,0,-0.05,width=0.015,color='red')
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.title("Grow the cluster of aligned spins")

np.random.seed(45)
plt.subplot(1, 3, 3)
c1 = plt.Circle((0, 0), 0.1, facecolor='w',edgecolor='k')
c2 = plt.Circle((0.25, 0), 0.1, facecolor='w',edgecolor='gray')
c3 = plt.Circle((-0.25, 0), 0.1, facecolor='w',edgecolor='gray')
c4 = plt.Circle((-0.25, 0.25), 0.1, facecolor='w',edgecolor='gray')
c5 = plt.Circle((0, -0.25), 0.1, facecolor='w',edgecolor='gray')
plt.gca().add_patch(c1);plt.gca().add_patch(c2);plt.gca().add_patch(c3);
plt.gca().add_patch(c4);plt.gca().add_patch(c5)
plt.grid(color='gray')
for axi in (plt.gca().xaxis, plt.gca().yaxis):
    for tic in axi.get_major_ticks():
        tic.tick1On = tic.tick2On = False
        tic.label1On = tic.label2On = False

for (x,y) in itertools.product(np.linspace(-1,1,9),repeat=2):
    if (x,y) in [(0,0),(0.25,0),(-0.25,0),(-0.25,0.25),(0,-0.25)]:
        plt.arrow(x,y-0.05,0,0.05,width=0.015,color='blue')
        continue
    if np.random.randint(2):
        plt.arrow(x,y-0.05,0,0.05,width=0.015,color='blue')
    else:
        plt.arrow(x,y+0.05,0,-0.05,width=0.015,color='red')
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.title("Flip the whole cluster")

plt.savefig("cluster_algorithm_fig1")
plt.show()
```
