# 1. Introduction of Ising model

In this notebook, we will be exploring Markov Chain Monte Carlo - a numerical method that helps with exploring the properties of thermal/statistical systems. An example of such a system is the Ising model, where we work with a square lattice in the plane, and we imagine (classical) spins living at each of the vertices that either point up or point down, as in the following picture:

```python
plt.figure(figsize=(8,8))
plt.grid(color='gray')

for axi in (plt.gca().xaxis, plt.gca().yaxis):
    for tic in axi.get_major_ticks():
        tic.tick1On = tic.tick2On = False
        tic.label1On = tic.label2On = False

for (x,y) in itertools.product(np.linspace(-1,1,9),repeat=2):
    if np.random.randint(2):
        plt.arrow(x,y-0.05,0,0.05,width=0.015,color='blue')
    else:
        plt.arrow(x,y+0.05,0,-0.05,width=0.015,color='red')
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.show()
```

The Hamiltonian (energy-function) is given by summing over the 'bonds' that live on the edges of the square lattice between the different spins, we can write this as:
$$H(\sigma) = -J \sum_{\langle i,j \rangle} \sigma_i \sigma_j $$
where $i,j$ are indexing the spins, $\sigma_i$ represents the spin (either +1 or -1) at that site, and the sum is over all the bonds in the lattice. Now that we've written down a phase-space and a Hamiltonian, we can consider the system at a constant temperature $T$, in which case each different configuration of spins is said to appear with probability:
$$P(\sigma) = \frac{1}{Z} e^{-\beta H(\sigma)}, \quad Z = \sum_{\sigma} e^{-\beta H(\sigma)}$$
where $Z$ (the partition function) is the normalization constant needed for the probabilities to be consistent. In all that follows, we will be working on finite lattices so we need not worry about technicalities arising in the continuum limit. Let's think about the interpretation of this Hamiltonian for a second. If $J > 0$, we say that the system is *ferromagnetic*, because the ground states (lowest energy) will be given by when all the spins point in the same direction, like in a traditional magnet. We see that the lower the energy, the higher the probability of observing that state in the thermal ensemble. Conversely, when $J < 0$ we see that the spins like to be opposite to each other, and we call the system *antiferromagnetic*

```python
plt.figure(figsize=(12,6))

plt.subplot(1, 2, 1)
plt.grid(color='gray')
for axi in (plt.gca().xaxis, plt.gca().yaxis):
    for tic in axi.get_major_ticks():
        tic.tick1On = tic.tick2On = False
        tic.label1On = tic.label2On = False

for (x,y) in itertools.product(np.linspace(-1,1,9),repeat=2):
    plt.arrow(x,y-0.05,0,0.05,width=0.015,color='blue')
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.title("Ferromagnetic ground state, J > 0")

plt.subplot(1, 2, 2)
plt.grid(color='gray')
for axi in (plt.gca().xaxis, plt.gca().yaxis):
    for tic in axi.get_major_ticks():
        tic.tick1On = tic.tick2On = False
        tic.label1On = tic.label2On = False

for (i,(x,y)) in enumerate(itertools.product(np.linspace(-1,1,9),repeat=2)):
    if i%2:
        plt.arrow(x,y-0.05,0,0.05,width=0.015,color='blue')
    else:
        plt.arrow(x,y+0.05,0,-0.05,width=0.015,color='red')
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.title("Antiferromagnetic ground state, J < 0")

plt.show()
```

Now that we have defined our thermal system, we might wonder how to extract physical quantities, such as the average energy, or the average magnetization. Formally, given any function $f(\sigma)$, we would like to be able to compute thermal averages:
$$\langle f \rangle := \frac{1}{Z} \sum_\sigma f(\sigma) e^{-\beta H(\sigma)}$$
The problem is that this is (naively) very computationally intractable! If we have a 50 x 50 lattice, then we are summing over 2500 spins, and since each spin can take two values, there are $2^{2500}$ terms in our sum! It seems that perhaps we could use symmetry arguments to cut this number down quite a bit, but it still poses a challenge! 

# 2. Introduction to Markov-Chain Monte-Carlo (MCMC)

In general, these kinds of thermal averages are very difficult to do: MCMC gives us a method of *approximating* the thermal average by carefully sampling from the space. Here, we will give a general introduction to MCMC by describing th Metropolis-Hastings algorithm, which is the easiest MCMC to implement. We will not give a proof of its correctness, for that we refer to the [wikipedia page](https://en.wikipedia.org/wiki/Metropolis%E2%80%93Hastings_algorithm) (which is actually pretty good). In the general case, suppose that we want to sample from a probability distribution $P(x)$ over a sample space $x \in \Omega$: the idea is that we will create a random walk through this space which will (asymptotically) reproduce this probability distribution. To run the algorithm, we will also need to write down a proposal distribution $g(x' | x)$, which will give the initial probability of jumping from state $x$ to state $x'$. It is necessary that this function $g$ allows you to actually explore the whole phase-space. Then:

 - Pick an initial state $x_0$ (asymptotically it doesn't matter what this is, but may give worse/better performance depending on what you pick)
 - Then, for each iteration, pick a new state $x'$ based on the old state $x_{n}$ via the proposal distribution $g(x' | x)$. For this proposal, calculate the acceptance probability $A(x',x_n) = \text{min}(1,\frac{P(x')}{P(x_n)}\frac{g(x_n|x')}{g(x'|x_n)})$. Generate a uniform random $u \in [0,1]$, if this random number is less than the acceptance probability, then set $x_{n+1} = x'$, else set $x_{n+1} = x_n$
 
 As an example, lets suppose that we want to sample from a gaussian distribution using metropolis-hastings. Then, we can use the following code:

```python
#Gaussian Distribution in x, with mean mu and s.d. sig
def gaussian(x, mu, sig):
    return (1./(np.sqrt(2.*np.pi)*sig))*np.exp(-np.power((x - mu)/sig, 2.)/2)

def metropolis_step(x):
    n = x + (np.random.random()-0.5)
    a = min(1,(gaussian(n,0,1)/gaussian(x,0,1)))
    if np.random.random() < a:
        return n
    return x

x = [0.2] #initialize at 0.2
for i in range(100000):
    x.append(metropolis_step(x[-1]))
```

```python
plt.figure(figsize=(8,8))
plt.title("Example of convergence to gaussian distribution")
bins = np.linspace(-3,3,50)
plt.hist(x[:50],bins=bins,density=1,label='50 datapoints',alpha=0.5)
plt.hist(x[:1000],bins=bins,density=1,label='1000 datapoints',alpha=0.5)
plt.hist(x[:100000],bins=bins,density=1,label='100000 datapoints',alpha=0.5)
plt.plot(np.linspace(-3,3,100),gaussian(np.linspace(-3,3,100),0,1),label='true distribution')
plt.legend()
plt.show()
```

As we see in the above plot, when we increase the number of samples, the MCMC eventually converges very nicely to the true distibution. Since we initialized the markov-chain at 0.2, we see that there is a positive bias that slowly fades over time - this is just to stress the importance of running the MCMC for a long-enough duration. What people will usually also do is to run the MCMC for a set number of steps, and throw out the first few hundred/thousand steps to allow the system to appropriately 'thermalize' (as we see in this plot, for the proposal function we used we should at least throw out a few hundred).

# 3. Applying MCMC to Ising Model!

Finally, let us turn our attention to applying the MCMC to extracting physical information out of the Ising Model. In this case, our sample space $\Omega$ is now the space of all spin assigments. We will choose a very proposal function $g(\sigma' | \sigma)$, where we pick a random spin, and flip it. Let us see what the average energy $\langle H \rangle$ is, for the ferromagnetic case $J > 0$ as a function of temperature. First, note that when computing the acceptance probability 
$$\frac{P(\sigma')}{P(\sigma)} = \frac{e^{-\beta H(\sigma')}}{e^{-\beta H(\sigma)}} = e^{-\beta (H(\sigma') - H(\sigma)}$$ 
the factors of $Z$ cancel out (which is good, because $Z$ is hard to come by), and the acceptance probability only depends on the change in the hamiltonian. Note that we can compute this purely locally if we are choosing a random spin to flip - this change in energy only depends on the values of the spins in the latttice that are connected to the spin we are changing. 

```python
def ising_hamiltonian(J,x):
    return -J*(np.sum(x*np.roll(x,1,axis=0)) + np.sum(x*np.roll(x,1,axis=1)))

def calc_staple(x,idx):
    nbhs = [((idx[0]+1)%x.shape[0],idx[1]),
            ((idx[0]-1)%x.shape[0],idx[1]),
            (idx[0],(idx[1]+1)%x.shape[1]),
            (idx[0],(idx[1]-1)%x.shape[1])]
    return sum(map(lambda y:x[y],nbhs))
    
# mutates x, measures delta(H)
def ising_metropolis_step_h(J,beta,x):
    idx = (np.random.randint(x.shape[0]),np.random.randint(x.shape[1]))
    staple = calc_staple(x,idx)
    if np.random.random() < np.exp(-2*beta*J*(x[idx]*staple)):
        x[idx] = -x[idx]
        return -2*J*x[idx]*staple
    return 0
```

```python
T_bins = np.linspace(0.1,6,100)
x = np.ones(shape=(25,25))
thermal_h = []
for T in np.linspace(0.1,6,100):
    #Burn 1000
    for i in range(1000):
        ising_metropolis_step_h(1,1/T,x)
    hist = [ising_hamiltonian(1,x)]
    for i in range(10000):
        hist.append(hist[-1] + ising_metropolis_step_h(1,1/T,x))
    thermal_h.append(np.average(hist))
thermal_h = np.array(thermal_h)/(25*25*2)
```

Before we check out what the average energy looks like, lets take a brief detour and figure out what we *expect* the answer to be. The nice thing about the Ising model is that we actually have an exact solution in the 2D square lattice, given by Onsager! In the thermodynamic limit, the free energy is given by the formula:
$$f := - \frac{\ln Z}{\beta} , \quad -\beta f = \ln 2 + \frac{1}{8 \pi^2} \int_0^{2 \pi} d \theta_1 \int_0^{2 \pi}d \theta_2 \ln \left[ \cosh (2 \beta J)^2 - \sinh (2 \beta J) (\cos (\theta_1) + \cos(\theta_2)) \right] $$
We can extract the thermal energy by taking an appropriate derivative of the free energy:
$$\langle H \rangle = -\frac{\partial \ln Z}{\partial \beta} = \frac{1}{8 \pi^2} \int_0^{2 \pi} d \theta_1 \int_0^{2 \pi}d \theta_2 \frac{2 J \cosh (2 \beta  J)
   \left(\cos \left(\theta
   _1\right)+\cos \left(\theta
   _2\right)-2 \sinh (2 \beta 
   J)\right)}{\cosh ^2(2 \beta 
   J)-\left(\cos \left(\theta
   _1\right)+\cos \left(\theta
   _2\right)\right) \sinh (2 \beta 
   J)}$$
We can numerically integrate this expression, and compare it to our MCMC results, which is what we have done in the plot below. 

```python
def exact_integrand(J,beta,theta1,theta2):
    return (2*J*np.cosh(2*J*beta)*(np.cos(theta1)+np.cos(theta2) - 2*np.sinh(2*J*(beta))))/\
            (np.cosh(2*J*beta)**2 - (np.cos(theta1)+np.cos(theta2))*np.sinh(2*J*beta))

exact_h = []
for T in T_bins:
    exact_h.append(dblquad(lambda y,x: exact_integrand(1,1/T,y,x),0,2*np.pi,0,2*np.pi))
exact_h = np.array(exact_h)/(8*np.pi**2)
```

```python
plt.figure(figsize=(8,8))
plt.title("Thermal average energy per bond")
plt.plot(T_bins,thermal_h,label="MCMC results,25 x 25 lattice")
plt.plot(T_bins,exact_h[:,0]/2,label="Exact continuum result")
plt.ylabel("energy per bond")
plt.xlabel("temperature")
plt.legend()
plt.show()
```

As we can see, we have great alignment! To physically interpret what's happening, notice that at low temperatures we should be in the ground state, so the average energy per bond should just be $-1$. As we raise the temperature, we introduce more thermal fluctuations to the system, and the energy per bond should slowly rise to $0$ as $t \to \infty$. Now, what if we turned our attention to something a little more spicy, the magnetisation? We expect there to be a phase transition in the 2D Ising Model, where below a critical temperature, the spins spontaneously magnetize. We will measure this by the absolute magnetisation variable:
$$\langle M \rangle = \sum_\sigma e^{-\beta H(\sigma)} \left| \sum_i \sigma_i \right|$$

```python
# Takes a while... 
def magnetisation(x):
    return np.abs(np.sum(x))

x = np.ones(shape=(25,25))
thermal_m = []
for T in np.linspace(1.5,3,25):
    #Burn 1000
    for i in range(2000):
        ising_metropolis_step_h(1,1/T,x)
    hist = [magnetisation(x)]
    for i in range(500000):
        ising_metropolis_step_h(1,1/T,x)
        hist.append(magnetisation(x))
    thermal_m.append(np.average(hist))
thermal_m = np.array(thermal_m)/(25*25)
```

```python
plt.figure(figsize=(8,8))
plt.title("Thermal average magnetisation per bond")
plt.scatter(np.linspace(1.5,3,25),thermal_m,label='MCMC results, 25 x 25 lattice')
plt.ylabel("magnetisation per bond")
plt.xlabel("temperature")
plt.axvline(2/np.log(1+np.sqrt(2)),label='exact critical temperature',color='red')
plt.legend()
plt.show()
```

As we can see, this plot is a lot messier, and we lose a lot of statistics as we crank up the temperature. In the limit that the lattice size goes to infinity, you expect there to be a sharp-cutoff in this diagram at the critical temperature. To get better plots of this kind, we can improve our methods by changing how we update our states, for instance using cluster algorithms such as the Wolff algorithm.
