# 1. Cluster Algorithms

In this notebook, we'll introduce a significantly better method for MCMC for spin systems such as the Ising model - known as the cluster algorithms. Up until now, we have been using algorithms that flip a single spin at every update step, and our algorithm looked something like this:

```python
def ising_hamiltonian(J,x):
    """
    Calculates the hamiltonian of a given configuration of spins
    Toroidal boundary conditions
    
    Inputs
      J : Strength of interaction
      x : 2D-numpy array of spins +-1
    """
    return -J*(np.sum(x*np.roll(x,1,axis=0)) + np.sum(x*np.roll(x,1,axis=1)))


def calc_staple(x,idx):
    """
    Calculates the 'staple' at a given site, which is defined as 
    the sum of the neighboring spins. 
    
    Inputs
      x : 2D-numpy array of spins, +-1
      idx : tuple (x,y) of indices for a given spin
    """
    nbhs = [((idx[0]+1)%x.shape[0],idx[1]),
            ((idx[0]-1)%x.shape[0],idx[1]),
            (idx[0],(idx[1]+1)%x.shape[1]),
            (idx[0],(idx[1]-1)%x.shape[1])]
    return sum(map(lambda y:x[y],nbhs))
    
def singleflip(J,beta,x):
    """
    Perform one MCMC step using the singleflip update
    
    Inputs
      J : Interaction strength
      beta : inverse temperature
      x : 2D-numpy array of spins, +-1. Is mutated. 
      
    Returns
      (change in energy of the configuration,
       change in the magnetisation)
    """
    idx = (np.random.randint(x.shape[0]),np.random.randint(x.shape[1]))
    staple = calc_staple(x,idx)
    if np.random.random() < np.exp(-2*beta*J*(x[idx]*staple)):
        x[idx] = -x[idx]
        return (-2*J*x[idx]*staple,2*x[idx])
    return (0,0)
```

Recall from our discussion about MCMC however, that in using the Metropolis-Hastings algorithm, we can choose different proposal distributions. For instance, in using the single flip algorithm, if we are in the regime $T < T_c$ where the temperature is cold, the spins will all form large clusters all pointing in the same direction. At any given MCMC step, there is an overwhelmingly small probabiility that the spin gets flipped (it likes to be the same as its neighbors), so the algorithm gets stuck for very long periods of time, without effectively exploring the phasespace. In this situation, its very difficult to get effective statistics on any physical quantities.

In this situation, we can swap to a different proposal (called *cluster algorithms*), where we propose to flip entire clusters of spins, instead of trying to flip individual spins. For high-temperature, this doesn't yield any improvement over single-spin (because the clusters will never be big because it just looks like random noise), however at temperatures near and below the critical temperature, these algorithms are necessary for effective simulation. 

The Wolff cluster algorithm (for ising model) proceeds as follows
  - Pick a random site to act as the seed. 
  - Consider all the bonds connected to the seed, if they are pointing in the same direction, 'activate' the bonds with probability $1 - e^{-2\beta J}$, and add them to the cluster. Proceed by looking at all the bonds connected to the cluster.
  - Flip the whole cluster.

The two conditions that are needed for a MCMC approach to model the correct (Gibbs) distribution are ergodicity and detailed balance. Ergodicity is the requirement that we can actually explore all of phase space - this is true because there is always some probability of flipping a single spin as a cluster, so we can go from any one configuration to any other by a sequence of single flips. We can check detailed balance by imagining flipping a $+1$ cluster to a $-1$ cluster to move from spin configuration $\sigma$ to $\sigma'$
$$\frac{P(\sigma \to \sigma')}{P(\sigma' \to \sigma)} = \frac{P(\text{All external bonds of cluster to $+1$ are not activated})}{P(\text{All external bonds of cluster to $-1$ are not activated})} = \frac{\prod_\text{external +1} e^{-2\beta J}}{\prod_\text{external -1} e^{-2 \beta J}} = e^{-\beta \Delta H}$$

```python
def wolffclusterflip(J,beta,x):
    """
    Perform one MCMC step using the Wolff algorithm (flip one cluster)
    
    Inputs
      J : Interaction strength
      beta : inverse temperature
      x : 2D-numpy array of spins, +-1. Is mutated. 
      
    Returns
      (change in energy of the configuration,
       change in magnetisation)
    """
    
    #Some convenience:
    dirs = {0: (0,1), 1: (1,0), 2: (0,-1), 3:(-1,0)}
    change = 0
    L = (x.shape[0],x.shape[1])
    
    #First, pick a random seed
    seed = (np.random.randint(x.shape[0]),np.random.randint(x.shape[1]))
    cluster = [seed]
    legs = [(seed,i) for i in range(4)]
    possible_breaks = []
    
    #Grow the cluster
    while len(legs) != 0:
        l = legs.pop()
        #print(l)
        target = ((l[0][0] + dirs[l[1]][0])%L[0], (l[0][1] + dirs[l[1]][1])%L[1])
        if target in cluster:
            continue
        if x[l[0]] != x[target]:
            change += 1
            continue
        if np.random.random() < 1 - np.exp(-2*beta*J):
            cluster.append(target)
            [legs.append((target,i)) for i in range(4) if i != (l[1]+2)%4]
        else:
            #Same, but bond not activated
            #It could be activated later on though, so we need to save 
            possible_breaks.append(target)
            #change -= 1
    
    for i in possible_breaks:
        if i not in cluster:
            change -= 1
    
    #Flip the cluster
    #print(cluster)
    for i in cluster:
        x[i] = -x[i]
    
    return (-2*J*change,2*len(cluster)*x[i])
```

We can check quickly that these two approaches give the same average energy $\langle H \rangle$ for the Ising model on a $25 \times 25$ lattice. From the previous notebook, we already know how to get the exact result:

```python
#Change this if you get bored of waiting for sims to finish running
T_bins = np.linspace(0.1,6,100)

#Using single-flip algorithm 
x = np.ones(shape=(25,25))
thermal_h_singleflip = []
for T in T_bins:
    #Burn 1000
    for i in range(1000):
        singleflip(1,1/T,x)
    hist = [ising_hamiltonian(1,x)]
    for i in range(10000):
        hist.append(hist[-1] + singleflip(1,1/T,x))
    thermal_h_singleflip.append(np.average(hist))
thermal_h_singleflip = np.array(thermal_h_singleflip)/(25*25*2)
```

```python
#Using cluster-flip algorithm 
x = np.ones(shape=(25,25))
thermal_h_clusterflip = []
for T in T_bins:
    #print(T)
    #Burn 1000
    steps = 0
    while steps < 100:
        steps += np.sqrt(np.abs(wolffclusterflip(1,1/T,x)[1]))
    #print("Done thermalising")
    hist = [ising_hamiltonian(1,x)]
    while steps < 1000:
        result = wolffclusterflip(1,1/T,x)
        hist.append(hist[-1] + result[0])
        steps += np.sqrt(np.abs(result[1]))
    thermal_h_clusterflip.append(np.average(hist))
thermal_h_clusterflip = np.array(thermal_h_clusterflip)/(25*25*2)
```

```python
#Exact result
def exact_integrand(J,beta,theta1,theta2):
    return (2*J*np.cosh(2*J*beta)*(np.cos(theta1)+np.cos(theta2) - 2*np.sinh(2*J*(beta))))/\
            (np.cosh(2*J*beta)**2 - (np.cos(theta1)+np.cos(theta2))*np.sinh(2*J*beta))
exact_h = []
for T in T_bins:
    exact_h.append(dblquad(lambda y,x: exact_integrand(1,1/T,y,x),0,2*np.pi,0,2*np.pi))
exact_h = np.array(exact_h)/(8*np.pi**2)
```

```python
plt.figure(figsize=(8,8))
plt.title("Thermal average energy per bond")
plt.plot(T_bins,thermal_h_singleflip,label="MCMC results with singleflip,25 x 25 lattice")
plt.plot(T_bins,thermal_h_clusterflip,label="MCMC results with clusters,25 x 25 lattice")
plt.plot(T_bins,exact_h[:,0]/2,label="Exact continuum result")
plt.ylabel("energy per bond")
plt.xlabel("temperature")
plt.legend()
plt.show()
```

As we can see, both methods reproduce the exact result of the average energy, as we expect. Note however that it is difficult to directly compare the two methods, because the cluster method takes more time per MCMC step to update (as it must grow the cluster, and only flips the cluster at the end). To do correct comparisons, perhaps the best way to compare the two methods is by time-testing them in other situations, which we will explore in the rest of this notebook.

# 2. Investigating the Phase Transition

Now, lets look at the more 'spicy' variable, the average magnetisation $\langle |\sum_i \sigma_i| \rangle_\text{Gibbs}$. This variable is sensitive to the phase transition, in the sense that in the thermodynamic limit $L \to \infty$, this is positive for $T < T_c$, and zero for $T > T_c$. To compare the two methods, lets try to allocate them the same amount of time to compute the magnetisaton as a function of $\beta$

```python
#Change this if you get bored of waiting for sims to finish running
T_bins = np.linspace(0.1,6,100)
```

```python
%%time 
#Using single-flip algorithm 
x = np.ones(shape=(25,25))
M_single = []
for T in T_bins:
    #Burn 1000
    for i in range(1000):
        singleflip(1,1/T,x)
    hist = [np.sum(x)]
    for i in range(10000):
        result = singleflip(1,1/T,x)
        hist.append(hist[-1] + result[1])
    M_single.append(np.average(np.abs(hist)))
M_single = np.array(M_single)/(25*25)
```

```python
%%time
#Using cluster-flip algorithm 
x = np.ones(shape=(25,25))
M_cluster = []
for T in T_bins:
    steps = 0
    while steps < 100:
        steps += np.sqrt(np.abs(wolffclusterflip(1,1/T,x)[1]))
    hist = [np.sum(x)]
    while steps < 1000:
        result = wolffclusterflip(1,1/T,x)
        hist.append(hist[-1] + result[1])
        steps += np.sqrt(np.abs(result[1]))
    M_cluster.append(np.average(np.abs(hist)))
M_cluster = np.array(M_cluster)/(25*25)
```

```python
plt.figure(figsize=(8,8))
plt.title("Thermal average energy per bond")
plt.plot(T_bins,M_single,label="Magnetisation, singleflip MCMC")
plt.plot(T_bins,M_cluster,label="Magnetisation, cluster MCMC")
plt.plot(T_bins,np.maximum(0,1-np.sinh(2/T_bins)**(-4))**(1/8),label="Exact continuum result")
plt.axvline(2/np.log(1+np.sqrt(2)),label='exact critical temperature',color='red')
plt.ylabel("energy per bond")
plt.xlabel("temperature")
plt.legend()
plt.show()
```

As we can see, we get much nicer, smoother results near the critical temperature, even though both methods ran for roughly the same total time. You might be wondering why they aren't matching up with the continuum result: that's because for finite lattice sizes (we are working with 25 x 25 lattices here) you will never have a phase transition, its only in the thermodynamic limit $L \to \infty$ that the shart cutoff shows up. 

Another way to think about whats happening near the critical temperature is by looking at the correlations between spins that are spatially very separated, $\langle \sigma_i \sigma_j \rangle_\text{Gibbs}$ where $i,j$ are sites that are separated. Using the cluster methods, we can investigate what this looks like, which we might do later.
